```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
#loading data file
df_sf = pd.read_csv('Datasets\Police_Department_Incidents.csv') 
```

```python
'''
Source:
https://www.latlong.net/place/san-francisco-ca-usa-594.html#:~:text=The%20latitude%20of%20San%20Francisco,%C2%B0%2025'%2052.6692''%20W.
San Francisco, CA, USA
Latitude and longitude coordinates are: 37.773972, -122.431297 

Let's check whether that the longitudes and latitues are within the san francisco area
.....
seems like we may need to scale a bit those values further on
'''
```

```python
#detecing location outliers
#Boxplots of latitud values by districts 
df_sf.boxplot(column = 'Y', by = 'PdDistrict', figsize=(20,20));
plt.title('Latitude Outliers by District')
```

```python
#Boxplots of longitude values by districts
df_sf.boxplot(column = 'X', by = 'PdDistrict', figsize=(25,25));
plt.title('Longitude Outliers by District')
```

```python
crime_category_names = df_sf['Category'].unique()
crime_categories = df_sf['Category'].value_counts().unique()

# Assign a color index to each crime category
color_map = dict(zip(crime_category_names, range(len(crime_categories))))

# Draw a scatter plot to show the crimes distribution
fig = plt.figure(figsize=(10,10))
plt.scatter(df_sf.X, df_sf.Y, c=df_sf.Category.map(color_map), cmap=plt.cm.gist_ncar, alpha = 0.01)
plt.title("SF Crime Locations", fontsize=12)
plt.xlabel('Longitude', fontsize=10)
plt.ylabel('Latitude', fontsize=10)
```

```python
'''
normalising latitud and longitud values 

X  -  Y 

''' 
```

```python
#dropping the outliers for the visualisation 
# putliers dropped
df_sf.drop(df_sf[df_sf['Y'] == df_sf['Y'].max()].index, inplace=True)
df_sf.drop(df_sf[df_sf['X'] == df_sf['X'].max()].index, inplace=True)
crime_category_names = df_sf['Category'].unique()
crime_categories = df_sf['Category'].value_counts().unique()

# Assign a color index to each crime category
color_map = dict(zip(crime_category_names, range(len(crime_categories))))

# Draw a scatter plot to show the crimes distribution
fig = plt.figure(figsize=(10,10))
plt.scatter(df_sf.X, df_sf.Y, c=df_sf.Category.map(color_map), cmap=plt.cm.gist_ncar, alpha = 0.01)
plt.title("SF Crime Locations", fontsize=12)
plt.xlabel('Longitude', fontsize=10)
plt.ylabel('Latitude', fontsize=10)
```

```python
#boxplots after dropping outliers in Y label
df_sf.boxplot(column = 'Y', by = 'PdDistrict', figsize=(20,20));
plt.title('Latitude Outliers by District')
```
