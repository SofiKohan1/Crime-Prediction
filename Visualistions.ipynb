```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import colors
```

```python
#loading data file
df_sf = pd.read_csv('Datasets\Police_Department_Incidents.csv')
```

```python
sns.pairplot(df_sf, hue='Category')
```

```python
#matplotlib asending orfder of crimes per category
fig=plt.figure(figsize=(15, 15))
df_sf_cats = df_sf['Category'].value_counts()
plt.title("\n Number of crimes per category\n")
plt.xlabel("\n Count")
plt.ylabel("Cateogry") 
ax = df_sf_cats.plot(kind='barh', color='purple')
```

```python
#seabor nvisualisation of crimes per category but couldnt order it :l
fig, ax = plt.subplots(figsize=(15,15))
sns.histplot(data=df_sf, y="Category", hue="Category")
```

```python
#number of crimes per district barchart with matplotbil

fig=plt.figure(figsize=(12, 12))
p_d = df_sf['PdDistrict'].value_counts()
plt.title("\n Number of crimes per district\n")
plt.xlabel("\n Total number of crimes")
plt.ylabel("District") 
ax = p_d.plot(kind='barh', color='purple')
```

```python
print(df_sf['Category'].unique())
```

```python
crime_category_names = df_sf['Category'].unique()
crime_categories = df_sf['Category'].value_counts().unique()

# Assign a color index to each crime category
color_map = dict(zip(crime_category_names, range(len(crime_categories))))

# Draw a scatter plot to show the crimes distribution
fig = plt.figure(figsize=(10,10))
plt.scatter(df_sf.X, df_sf.Y, c=df_sf.Category.map(color_map), cmap=plt.cm.gist_ncar, alpha = 0.01)
plt.title("SF Crime Locations", fontsize=12)
plt.xlabel('Longitude', fontsize=10)
plt.ylabel('Latitude', fontsize=10)
```

```python
#heatmap between cateogry and districs
#select and store the features we want to view in heatmap

categories = df_sf['Category'].unique()
districts = df_sf['PdDistrict'].unique()
data = []
for cat in categories:
    get_data = []
    for distr in districts:
        get_data.append(len(df_sf[(df_sf['PdDistrict'] == distr) & (df_sf['Category'] == cat)]))
    data.append(get_data)

fig, ax = plt.subplots(1, 1, figsize =(10,15))
heat_df = pd.DataFrame(data, index=categories, columns=districts)
sns.heatmap(heat_df, cmap='YlGnBu', linewidths=.2)
```

```python
# Let's drop the outlier for now, since it is skewing our visualization
df_sf.drop(df_sf[df_sf['Y'] == df_sf['Y'].max()].index, inplace=True)
df_sf.drop(df_sf[df_sf['X'] == df_sf['X'].max()].index, inplace=True)


crime_category_names = df_sf['Category'].unique()
crime_categories = df_sf['Category'].value_counts().unique()

# Assign a color index to each crime category
color_map = dict(zip(crime_category_names, range(len(crime_categories))))

# Draw a scatter plot to show the crimes distribution
fig = plt.figure(figsize=(10,10))
plt.scatter(df_sf.X, df_sf.Y, c=df_sf.Category.map(color_map), cmap=plt.cm.gist_ncar, alpha = 0.01)
plt.title("SF Crime Locations", fontsize=12)
plt.xlabel('Longitude', fontsize=10)
plt.ylabel('Latitude', fontsize=10)

```
